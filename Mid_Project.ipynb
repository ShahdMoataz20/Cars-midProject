# **Cars Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/vehicles.csv', engine= 'python', on_bad_lines='skip')
```

```python
df.shape
```

```python
df
```

```python
df.drop(['county'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['VIN']=df['VIN'].fillna(df['VIN'].mode()[0])
df['cylinders']=df['cylinders'].fillna(df['cylinders'].mode()[0])
df['drive']=df['drive'].fillna(df['drive'].mode()[0])
df['fuel']=df['fuel'].fillna(df['fuel'].mode()[0])
df['title_status']=df['title_status'].fillna(df['title_status'].mode()[0])
df['transmission']=df['transmission'].fillna(df['transmission'].mode()[0])
df['VIN']=df['VIN'].fillna(df['VIN'].mode()[0])
df['fuel'] = df['fuel'].fillna(df['fuel'].mode()[0])
df['odometer'] = df['odometer'].fillna(df['odometer'].mean())
df['paint_color'] = df['paint_color'].fillna(df['paint_color'].mode()[0])
df['condition'] = df['condition'].fillna(df['condition'].mode()[0])
df['manufacturer'] = df['manufacturer'].fillna(df['manufacturer'].mode()[0])
df['model'] = df['model'].fillna(df['model'].mode()[0])
df['year'] = df['year'].fillna(df['year'].mean())
df['type'] = df['type'].fillna(df['type'].mode()[0])
df['state'] = df['state'].fillna(df['state'].mode()[0])
df['size'] = df['size'].fillna(df['size'].mode()[0])
df['lat'] = df['lat'].fillna(df['lat'].mean())
df['long'] = df['long'].fillna(df['long'].mean())
df['price'] = df['price'].fillna(df['price'].mean())
df['image_url'] = df['image_url'].fillna(df['image_url'].mode()[0])
df['description'] = df['description'].fillna(df['description'].mode()[0])
df['posting_date'] = df['posting_date'].fillna(df['posting_date'].mode()[0])
```

# **EDA**

## **1- What are the most common car brands exist?**



```python
brand_counts = df['manufacturer'].value_counts().head(8)
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(x=brand_counts.values, y=brand_counts.index, palette='inferno')
plt.xlabel('Number of Cars')
plt.ylabel('Car Brand')
plt.title('Top 8 Most Common Car Brands in the Dataset')
```

# **2- What is the average price of cars by brand?**

```python
avg_price_by_brand = df.groupby('manufacturer')['price'].mean().sort_values(ascending=False).head(10)
avg_price_by_brand = avg_price_by_brand.reset_index()
```

```python
plt.figure(figsize=(8, 6))
sns.pointplot(x='price', y='manufacturer', data=avg_price_by_brand, color='Purple', markers='o', linestyles='-')

plt.xlabel('Average Price (USD)')
plt.ylabel('Car Brand')
plt.title('Top 10 Car Brands by Average Price')
```

# **3- What are the most common fuel types?**

```python
fuel_counts = df['fuel'].value_counts()
fuel_counts
```

```python
avg_price_by_fuel = df.groupby('fuel')['price'].mean().sort_values(ascending=False)
avg_price_by_fuel = avg_price_by_fuel.reset_index()
avg_price_by_fuel
```

```python
sns.violinplot(x='fuel', y='price', data=avg_price_by_fuel, palette='muted', order=df.groupby('fuel')['price'].median().sort_values().index)
```

# **4- What is the relationship between price and odometer readings?**

```python
df_filtered = df[(df['price'] > 500) & (df['price'] < 100000)].head(100)
correlation = df_filtered[['price', 'odometer']].corr()

plt.figure(figsize=(10, 6))
plt.hexbin(df_filtered['odometer'], df_filtered['price'], gridsize=50, cmap='Blues', mincnt=1)
plt.colorbar(label='Count')
plt.title('Hexbin Plot: Price vs. Odometer Reading', fontsize=14)
plt.xlabel('Odometer Reading (miles)', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```
